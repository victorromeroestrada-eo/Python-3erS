#3.- Programa en Python que lea 2 números y realice las 4 operaciones básicas

```python
num = float(input("Escriba un número: "))
num2 = float(input("Escriba otro: "))
op = num + num2
print (f"La suma es {op},", end=" ")
op = num - num2
print (f"la resta es {op},", end=" ")
op = num * num2
print (f"el producto es {op}", end=" ")
op = num / num2
print (f"y la división es {op}")
```

#4.- Programa en Python Que sirva para calcular el área de un triángulo

#5.- Programa en Python que lea una medida en centímetros y desplegar la misma medida pero convertida en pies y pulgadas

```python
cm = float(input("Escriba una medida de centímetros:"))
inch = cm / 2.54
ft = cm / 30.48
print (f"{cm} centímetros son {inch} pulgadas o {ft} pies.")
```

#6.- Programa en Python que lea una temperatura en grados centígrados y desplegarla en Grados Fahrenheit y grados kelvin

```python
cel = float(input("Escriba una medida de grados C°: "))
far = (cel * (9/5))+32
kel = cel + 273.15
print (f"{cel} °C equivale a {far} °F y {kel} K.")
```
